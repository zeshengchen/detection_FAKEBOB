# Code to plot CDF of nth-MEH-FEST detection results against nth-FAKEBOB countermeasures

```python
import numpy as np
import matplotlib.pyplot as plt

def preprocess(detections):
    print("size = %d, max = %f, min = %f, mean = %10.8f, std = %10.8f, D = %10.8f" % 
          (len(detections), np.max(detections),
           np.min(detections), np.mean(detections),
           np.std(detections), np.mean(detections)+3*np.std(detections)))
    
    index = np.zeros(len(detections))
    results = np.zeros(len(detections))    
    for i, result in enumerate(detections):
        index[i] = i + 1
        results[i] = result
    
    index = index / len(detections)
    results = results * 10000
    
    return index, results
    
def plot_cdf_original_audio_E2(train_data, test_data,
                               all_data_E2, all_data_E, file_name):
  
    train_index, train_results = preprocess(train_data)
    test_index, test_results = preprocess(test_data)
    all_index_E2, all_results_E2 = preprocess(all_data_E2)
    all_index_E, all_results_E = preprocess(all_data_E)
        
    plt.plot(train_results, train_index, 'r-', label='Training data ($E_2$)')
    plt.plot(test_results, test_index, 'b:', label='Test data ($E_2$)')
    plt.plot(all_results_E2, all_index_E2, 'k--', label='All original audios ($E_2$)')
    plt.plot(all_results_E, all_index_E, 'g-.', label="All original audios ($E$)")
    
    plt.xlabel(r'Energy in high frequencies (i.e., $E$ or $E_2$) $\mathrm{x 10^{4}}$', fontsize=12)
    plt.ylabel('CDF', fontsize=12)
    plt.tick_params(axis='both', labelsize=12)
    plt.legend()
    plt.tight_layout()
    
    plt.savefig(file_name, dpi = 300)
    plt.show()
    
    plt.close()

def plot_cdf_gmm_adv_audio_E2(illegal_original_detections_E,
                              all_original_detections_E2,
                              adv_epsilon_002_detections_E,
                              adv_epsilon_002_detections_E2,
                              file_name):
  
    org_index_E, org_results_E = preprocess(illegal_original_detections_E)
    org_index_E2, org_results_E2 = preprocess(all_original_detections_E2)
    adv_002_index_E, adv_002_results_E = preprocess(adv_epsilon_002_detections_E)
    adv_002_index_E2, adv_002_results_E2 = preprocess(adv_epsilon_002_detections_E2)
             
    plt.plot(org_results_E, org_index_E, 'k--', 
             label='Original illegal audios ($E$)')
    plt.plot(adv_002_results_E, adv_002_index_E, 'r:', 
             label='1st FAKEBOB audios ($E$)')
    plt.plot(org_results_E2, org_index_E2, 'b-.', 
             label='All original audios ($E_2$)')
    plt.plot(adv_002_results_E2, adv_002_index_E2, 'g-', 
             label='1st FAKEBOB audios ($E_2$)')
    
    plt.xscale('log')
    plt.xlabel(r'Energy in high frequencies (i.e., $E$ or $E_2$) $\mathrm{x 10^{4}}$', fontsize=12)
    plt.ylabel('CDF', fontsize=12)
    plt.tick_params(axis='both', labelsize=12)
    plt.legend()
    plt.tight_layout()
    
    plt.savefig(file_name, dpi = 300)
    plt.show()
    
    plt.close()
    
def plot_cdf_gmm_adv_audio_E11(illegal_original_detections_E10,
                               all_original_detections_E11,
                               adv_epsilon_002_detections_E10,
                               adv_epsilon_002_detections_E11,
                               file_name):
  
    org_index_E10, org_results_E10 = preprocess(illegal_original_detections_E10)
    org_index_E11, org_results_E11 = preprocess(all_original_detections_E11)
    adv_002_index_E10, adv_002_results_E10 = preprocess(adv_epsilon_002_detections_E10)
    adv_002_index_E11, adv_002_results_E11 = preprocess(adv_epsilon_002_detections_E11)
             
    plt.plot(org_results_E10, org_index_E10, 'k--', 
             label='Original illegal audios ($E_{10}$)')
    plt.plot(adv_002_results_E10, adv_002_index_E10, 'r:', 
             label='10th FAKEBOB audios ($E_{10}$)')
    plt.plot(org_results_E11, org_index_E11, 'b-.', 
             label='All original audios ($E_{11}$)')
    plt.plot(adv_002_results_E11, adv_002_index_E11, 'g-', 
             label='10th FAKEBOB audios ($E_{11}$)')
    
    plt.xscale('log')
    plt.xlabel(r'Energy in high frequencies (i.e., $E_{10}$ or $E_{11}$) $\mathrm{x 10^{4}}$', fontsize=12)
    plt.ylabel('CDF', fontsize=12)
    plt.tick_params(axis='both', labelsize=12)
    plt.legend()
    plt.tight_layout()
    
    plt.savefig(file_name, dpi = 300)
    plt.show()
    
    plt.close()

def plot_cdf_gmm_adv_audio_E11_E21(all_original_detections_E11,
                                   all_original_detections_E21,
                                   adv_epsilon_002_detections_E11,
                                   adv_epsilon_002_detections_E21,
                                   file_name):
    
    org_index_E11, org_results_E11 = preprocess(all_original_detections_E11)
    org_index_E21, org_results_E21 = preprocess(all_original_detections_E21)
    adv_002_index_E11, adv_002_results_E11 = preprocess(adv_epsilon_002_detections_E11)
    adv_002_index_E21, adv_002_results_E21 = preprocess(adv_epsilon_002_detections_E21)

    plt.plot(org_results_E11, org_index_E11, 'k--', 
             label='All original audios ($E_{11}$)')
    plt.plot(org_results_E21, org_index_E21, 'r:', 
             label='All original audios ($E_{21}$)')
    plt.plot(adv_002_results_E11, adv_002_index_E11, 'b-.', 
             label='10th FAKEBOB audios ($E_{11}$)')
    plt.plot(adv_002_results_E21, adv_002_index_E21, 'g-', 
             label='20th FAKEBOB audios ($E_{21}$)')
    
    plt.xscale('log')
    plt.xlabel(r'Energy in high frequencies (i.e., $E_{11}$ or $E_{21}$) $\mathrm{x 10^{4}}$', fontsize=12)
    plt.ylabel('CDF', fontsize=12)
    plt.tick_params(axis='both', labelsize=12)
    plt.legend()
    plt.tight_layout()
    
    plt.savefig(file_name, dpi = 300)
    plt.show()
    
    plt.close()    
    
def split_train_test(data, test_ratio):
    # the code refers to https://github.com/ageron/handson-ml2/blob/master/02_end_to_end_machine_learning_project.ipynb
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data[train_indices], data[test_indices]
```
