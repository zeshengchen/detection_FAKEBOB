# Code to evaluate the performance of (n+1)th-MEH-FEST detection method against nth-FAKEBOB countermeasures

```python
import numpy as np
import matplotlib.pyplot as plt

def get_data_info(data, name):
    print("name = %s, size = %d, max = %f, min = %f, mean = %10.8f, std = %10.8f, D = %10.8f" % 
          (name, len(data), np.max(data), np.min(data), np.mean(data), np.std(data), 
           np.mean(data)+3*np.std(data)))

def get_threshold(train_data):
    threshold = np.mean(train_data)+3*np.std(train_data)
    print("Threshold = %10.8f" % (threshold))
    return threshold
            
def split_train_test(data, test_ratio):
    # the code refers to https://github.com/ageron/handson-ml2/blob/master/02_end_to_end_machine_learning_project.ipynb
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data[train_indices], data[test_indices]

def get_train_test_data(illegal_audio, legal_audio):
    # put illegal and legal audio into one dataset
    org_audio = np.concatenate((illegal_audio, legal_audio))
    org_audio.sort()
    
    # split orginal audio into training and test data sets
    np.random.seed(42)
    train_data, test_data = split_train_test(org_audio, 0.2)
    
    return train_data, test_data    

def evaluate(illegal_audio, legal_audio, adv_audio, train_data, test_data):
    
    adv_audio = np.array(adv_audio)
    
    # show data info
    get_data_info(np.array(illegal_audio), 'Illegal audio')
    get_data_info(np.array(legal_audio), 'Legal audio')
    get_data_info(adv_audio, 'Adversarial audio')
    print('')    
    
    get_data_info(train_data, 'Training')
    get_data_info(test_data, 'Test')
    print('')

    D = get_threshold(train_data)
    
    num_fp = np.argwhere(test_data > D).flatten().size    
    num_fn = np.argwhere(adv_audio <= D).flatten().size 
    
    print("size of test data = %d, num false postivie = %d, FPR = %f" % 
         (test_data.size, num_fp, num_fp/test_data.size))
    print("size of adv audio = %d, num false negative = %d, FNR = %f" %
         (adv_audio.size, num_fn, num_fn/adv_audio.size))

def evaluate_with_new_threshold(test_data, adv_data, D):
    adv_data = np.array(adv_data)
    num_fp = np.argwhere(test_data > D).flatten().size    
    num_fn = np.argwhere(adv_data <= D).flatten().size 
    
    print("\nNew D = %f" % (D))
    print("size of test data = %d, num false postivie = %d, FPR = %f" % 
         (test_data.size, num_fp, num_fp/test_data.size))
    print("size of adv audio = %d, num false negative = %d, FNR = %f" %
         (adv_data.size, num_fn, num_fn/adv_data.size))
    
```
